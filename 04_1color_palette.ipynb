```python
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline 
import warnings #忽略警告
warnings.filterwarnings('ignore')
                        
current_palette = sns.color_palette() 
sns.palplot(current_palette)
```

```python
#seaborn提供的6種調色板 
theme_list=['deep', 'muted', 'pastel','bright', 'dark', 'colorblind'] 
for i in theme_list:  #逐一畫出不同color_palette
    sns.palplot(sns.color_palette(i))

```

```python
sns.color_palette("hls", 25) #自定義多色調色盤
```

```python
#隨機生成資料，並使用多色調色盤定義顏色
data = np.random.normal(size=(20, 12)) + np.arange(12) / 2 
sns.boxplot(data=data,palette=sns.color_palette("hls", 12))
```

```python
#單色漸層，default為淺到深
sns.palplot(sns.color_palette("Blues")) 
sns.palplot(sns.color_palette("Reds_r")) 
sns.palplot(sns.color_palette("Greens_r")) #_r表示為reverse
```

```python
#使用默認亮度和飽和度 
sns.palplot(sns.hls_palette(3))
```

```python
sns.palplot(sns.hls_palette(3, l=0.8, s=0.65)) #l:亮度，s:飽和度
```

```python
#lw表示線寬
plt.plot([0, 1], [0, 1], sns.xkcd_rgb["purple"], lw=2) #(0,0)畫到(1,1)
plt.plot([0, 1], [0, 2], sns.xkcd_rgb["green"], lw=3) #(0,0)畫到(1,2)
plt.plot([0, 1], [0, 3], sns.xkcd_rgb["orange"], lw=4) #(0,0)畫到(1,3)
```

```python
sns.palplot(sns.color_palette()) 
sns.palplot(sns.color_palette("cubehelix"))
```

```python
sns.palplot(sns.cubehelix_palette(6, start=0.8, rot=-0.5)) #線性變化幾格
```

```python
sns.palplot(sns.light_palette("green")) 
sns.palplot(sns.dark_palette("green"))
```

```python
#等高線
x, y = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=600).T #
pal = sns.dark_palette("green", as_cmap=True) 
sns.kdeplot(x, y, cmap=pal);
```

```python
#隨機等高面
x, y = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=600).T 
pal = sns.cubehelix_palette(light=1, as_cmap=True) 
sns.kdeplot(x, y, cmap=pal, shade=True)
```
